**PLOTS**

```python
plt.bar(df1['Region'], df1[' Estimated Unemployment Rate (%)'],color="maroon")
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('Unemployment Rate')
plt.title('State-wise Unemployement Rate')
plt.show()
```

```python
plt.bar(df1['Region'], df1[' Estimated Labour Participation Rate (%)'],color="green")
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('Labour Participation Rate')
plt.title('State-wise Labour Participation Rate ')
plt.show()
```

```python
sns.pairplot(df1)
```

```python
sns.lineplot(y=' Estimated Unemployment Rate (%)', x=' Date', data=df1,color="black")
plt.title('Date wise Unemployment Rate ')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
sns.lineplot(y=' Estimated Employed', x=' Date', data=df1)
plt.title('Date-wise Employment Rate')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Employment Rate')
plt.show()
```

```python
sns.lineplot(y=' Estimated Labour Participation Rate (%)', x=' Date', data=df1,color="black")
plt.title('Date wise Labour Participation Rate ')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Labour Participation Rate')
plt.show()
```

```python
sns.histplot(x=' Estimated Unemployment Rate (%)', data=df1, hue='Area',color="red")
plt.title('Area wise Unemployment Rate')
plt.xlabel('Unemployment Rate')
plt.show()
```

```python
sns.histplot(x=' Estimated Labour Participation Rate (%)', data=df1, kde=True, hue='Area')
plt.title('Area wise Labour Participation Rate')
plt.xlabel('Labour Participation Rate')
plt.show()
```

```python
sns.histplot(x=' Estimated Employed', data=df1, kde=True, hue='Area')
plt.title('Area-wise Employment Rate')
plt.xlabel('Employment Rate')
plt.show()
```

```python
sns.lineplot(y=' Estimated Employed', x=' Date', data=df1)
plt.title('Datewise Employment Rate')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Employment Rate')
plt.show()
```
